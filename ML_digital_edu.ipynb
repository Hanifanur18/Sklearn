# **Data preparation**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_url_train = 'https://raw.githubusercontent.com/Hanifanur18/Sklearn/main/train.csv'
file_url_test = 'https://raw.githubusercontent.com/Hanifanur18/Sklearn/main/test.csv'
df_train = pd.read_csv(file_url_train)
df_test = pd.read_csv(file_url_test)
```

```python
df_train.info()
```

```python
df_train = df_train.dropna()
df_train.info()
```

```python
df_train.head()
```

```python
df_train['education_form'].value_counts()
```

```python
df_train.groupby(by = 'education_form')['result'].sum()
```

```python
df_train['education_status'].value_counts()
```

```python
df_train.groupby(by = 'education_status')['result'].sum()
```

Define

1. feature (X)
2. target (y)
3. feature_test (X_test)

```python
X_train = df_train[['sex','graduation','education_status','education_form']]
X_train
```

```python
y_train = df_train['result']
y_train.shape
```

```python
X_test = df_test[['sex','graduation','education_status','education_form']]
X_test
```

data preprocessing

```python
from sklearn.preprocessing import OrdinalEncoder
ed = OrdinalEncoder(categories='auto')
X_train = ed.fit_transform(X_train)
X_train
```

```python
X_test = ed.fit_transform(X_test)
X_test
```

# **Create machine learning model using KNN**

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred_train = model.predict(X_train)
acc = accuracy_score(y_train, y_pred_train)
print(f'Accuracy :{acc}')
```

# **Result for test dataset**

```python
y_pred_test = model.predict(X_test)
df_test['result'] = y_pred_test
df_test
```
